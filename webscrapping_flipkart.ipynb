```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
from urllib.request import urlopen as uReq
import numpy as np
```

```python
k=[]
v=[]
rate=[]
i=1
searchString='samsung'
flipkart_url = "https://www.flipkart.com/search?q=" + searchString
uClient = uReq(flipkart_url)
flipkartPage = uClient.read()
#uClient.close()
bs=BeautifulSoup(flipkartPage,"html.parser")
pg=bs.find("div",{"class":"_2zg3yZ"})
page_no=int(pg.find("span").text[-2:])
while i<page_no:
    a=bs.find("a",{"class":"_3fVaIS"})
    #print(a)
    nextpage=a['href']
    #print(nextpage)
    productLink = "https://www.flipkart.com" + nextpage
    next_page_link=productLink.replace(productLink[-1],str(i+1))
    #print(next_page_link)
    #print(productLink)
    prodRes = requests.get(next_page_link)
    prodRes.encoding = 'utf-8'
    prod_html = BeautifulSoup(prodRes.text, "html.parser")
    products=prod_html.find_all("div",{"class":"_1UoZlX"})
    
    for product in products:
        name=str(product.find("div",{"class": "_3wU53n"}).text.strip())
        k.append(name)
        price=product.find("div",{"class": "_1vC4OE _2rQ-NK"}).text.replace('₹','').strip()
        v.append(price)
        try:            
            rating=product.find("div",{"class": "hGSR34"}).text.strip()
        except:
            rating='N/A'
        rate.append(rating)
        print("name ={}  price = {}  ".format(name,price))
        
        #print("name ={}  price = {}  ".format(name,price))
             
   
    i+=1
    
dictOfdata = {'name':k,'price':v,'rating':rate}
data=pd.DataFrame.from_dict(dictOfdata)
```
